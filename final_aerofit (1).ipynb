Many customers who buy the cheaper products KP281 and KP481 might be interested in upgrading to the premium KP781 if they use their products more or want better features. We should focus on identifying those customers who are using their products moderately or heavily and offer them special deals or information to encourage them to buy the KP781. This way, we can increase sales of the higher-priced product.


2. Marketing Should Match Customer Types
We should tailor our marketing messages depending on who the customers are:

For KP281, most buyers are low-income and often women. So, ads should focus on how affordable and useful the product is. Local stores and offline marketing can work well here.

For KP481, buyers are middle-income with balanced usage. We can use online ads and social media that talk about the product’s good value for the price.

For KP781, customers are usually high-income and highly educated, often men who use the product a lot. We should advertise through fitness influencers, online fitness groups, or premium gyms where serious users will notice it.


3. Bundle Products to Add Value
We can increase sales by offering bundled packages. For example, sell KP481 or KP781 together with useful accessories or service plans, so customers feel they are getting more for their money. For KP281, we can create family packs or deals for gyms that need multiple units at a lower cost.


4. Use Loyalty and Referral Programs
We should create programs to keep our current customers happy and bring in new ones:

For KP781 users, who are our most loyal and heavy users, a referral program can encourage them to tell friends and family about the product.

For KP481 users, a loyalty program with rewards or discounts can encourage them to stay with us longer and eventually upgrade.


5. Improve Products Based on Feedback
It’s important to listen to what customers say, especially KP781 users who expect premium quality. We can use their feedback to improve the product’s features and fix any problems. Also, to fill the gap between KP481 and KP781, we can think about making a new product priced between these two for customers who want something better than KP481 but can’t afford KP781 yet.


6. Focus on Right Sales Channels
Where we sell each product matters:

KP281 sells best in smaller cities or towns, so we should focus on local retail shops and markets there.

KP781 customers might be mostly in big cities and prefer to buy online or from premium gyms and stores, so we should strengthen our online presence and partnerships in metro areas.
